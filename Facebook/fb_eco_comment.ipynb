```python
# 點進留言後再打開'查看更多' & '所有留言'
def Expand_Post_Comment():
    
    time.sleep(random.randint(5, 10))
    # element_inside_popup = driver.find_element(By.XPATH, '//*[@id="facebook"]')
    # element_inside_popup.send_keys(Keys.END)
        
    
    # 【一、展開留言】 # 點開文章留言, 進去後最相關留言 => 所有留言
    comment_button = driver.find_element(By.XPATH, '//span[contains(text(), "留言")]')
    time.sleep(3)
    # action = ActionChains(driver)
    # action.move_to_element(comment_button).click().perform()
    comment_button.click()
    time.sleep(random.randint(5, 10))
    
    related_button = driver.find_element(By.XPATH, '//span[contains(text(), "最相關留言")]')
    time.sleep(3)
    related_button.click()
    time.sleep(random.randint(5, 10))
    
    related_button = driver.find_element(By.XPATH, '//span[contains(text(), "所有留言")]')
    time.sleep(3)
    related_button.click()
    time.sleep(random.randint(5, 10))


    # 【二、展開'顯示更多'】 # 點開留言裡面的顯示更多. 抓取完整文章內容
    show_more_buttons = driver.find_elements(By.XPATH, "//div[contains(text(), '顯示更多')]")

    if show_more_buttons:
        for button in show_more_buttons:
            try:
                button.click()
                time.sleep(random.randint(5, 10))
                print('Clicked "顯示更多" button')
            except:
                pass
            
            
        # 【三、模擬鍵盤按鍵，向下滾動彈出視窗的內容】  # selenium python how to scroll down in a pop up window
            element_inside_popup = driver.find_element(By.XPATH, '//*[@id="facebook"]')
            element_inside_popup.send_keys(Keys.END)
        
        
        # 【四、轉換html物件】    
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        

        #【五、找到Po文時間/Po文/留言時間/留言】
        
        article_times= soup.findAll("a", {"class":"x1i10hfl xjbqb8w x6umtig x1b1mbwd xaqea5y xav7gou x9f619 x1ypdohk xt0psk2 xe8uvvx xdj266r x11i5rnm xat24cr x1mh8g0r xexx8yu x4uap5 x18d9i69 xkhd6sd x16tdsg8 x1hl2dhg xggy1nq x1a2a7pz x1heor9g xt0b8zv xo1l8bm"})
        title = soup.find("div", {"class":"x1iorvi4 x1pi30zi x1swvt13 xjkvuk6"})
        print("一共:" + str(len(title)) + " 則文章...")
        
        times = soup.findAll("a", {"x1i10hfl xjbqb8w x6umtig x1b1mbwd xaqea5y xav7gou x9f619 x1ypdohk xt0psk2 xe8uvvx xdj266r x11i5rnm xat24cr x1mh8g0r xexx8yu x4uap5 x18d9i69 xkhd6sd x16tdsg8 x1hl2dhg xggy1nq x1a2a7pz xt0b8zv xi81zsa xo1l8bm"})        
        comments = soup.findAll("div", {"class": "xdj266r x11i5rnm xat24cr x1mh8g0r x1vvkbs"})
        print("一共:" + str(len(comments)) + " 則留言...")
    
        # 先設置寫進csv
        with open('posts_comment.csv', mode='w', encoding='utf-8', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(['文章', '時間', '留言時間', '留言'])
            
            # 只有一個結果, for in 
            for article_time in article_times:
                article_time_text = article_time.text
                title_text = title.text
                print("文章:", title_text)
                print("文章時間:",article_time_text)
                writer.writerow(['文章:', title_text])
                writer.writerow(['文章時間:', article_time_text])
                
                # 預期很多結果, for in range(len()) 
                for t in range(len(times)):
                    time_text = times[t].text
                    comment_text = comments[t].text
                    print("留言時間:", time_text)
                    print("留言:", comment_text)
                    writer.writerow(['留言時間:', time_text])
                    writer.writerow(['留言:', comment_text])
            writer.writerow(['已寫入CSV文件'])
            print("#################################")
                
        
            # 【四、關閉留言】
            # close_button = driver.find_element(By.CSS_SELECTOR, 'div.x1i10hfl div [aria-label]'
            try:
                close_button = driver.find_element(By.CSS_SELECTOR, 'div.x1i10hfl div')
                actions = ActionChains(driver)
                actions.move_to_element(close_button).click().perform()
            except:
                pass
                print("失敗")
    

# Expand_Post_Comment()
```
