```python
import pandas as pd
import datetime
```

```python
data_o = 'C:/Users/catal/Desktop/py/7-panda/ireland_wind.txt'
data = pd.read_csv(data_o, sep = '\s+', parse_dates = [[0, 1, 2]])
data.head(3)

# data
# 出來結果是 2070 ~ 2072、1973 ~ 1978
```

```python
# 調整跑掉的日期
def fix(x):
    if x.year >= 2070:
        year = x.year - 100
    else:
        year = x.year
    return datetime.datetime(year, x.month, x.day)

data['Yr_Mo_Dy'] = data['Yr_Mo_Dy'].apply(fix)
data 

# data.to_csv('output.csv', index=False)
```

```python
# # apply 練習
# data.apply(代入def名稱)
# 前面x是一個參數, 輸入每一個data值
import pandas as pd

data_apply = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

def multiply_by_two(x):
    return x * 2

result = data_apply.apply(multiply_by_two)

print(result)
```

```python
# 日期設為index
data.index = pd.to_datetime(data['Yr_Mo_Dy'])
data
```

```python
# 取 RPT 前十高
top_10_rpt = data.sort_values('RPT', ascending=False).head(10)
top_10_rpt
```

```python
# 取 RPT 前十高_2
top_10_rpt = data.sort_values('RPT', ascending=False).head(10)['RPT']
top_10_rpt
```

```python
# 取 RPT 總平均
rpt_mean = round(data['RPT'].mean(),2)
rpt_mean
```

```python
# 取 詳細統計資料
data.describe()
```

```python
# 取 所有城市 min, max, mean, std
day_st = pd.DataFrame()
day_st['min'] = data.drop('Yr_Mo_Dy', axis=1).min(axis=1)
day_st['max'] = data.drop('Yr_Mo_Dy', axis=1).max(axis=1)
day_st['mean'] = data.drop('Yr_Mo_Dy', axis=1).mean(axis=1)
day_st['std'] = data.drop('Yr_Mo_Dy', axis=1).std(axis=1)
day_st
```

```python
# 每個城市 六月平均
data.loc[data.index.month == 6].mean()
```

```python
# 每個城市 四~六月平均
mean_4 = data.loc[data.index.month == 4].mean()
mean_5 = data.loc[data.index.month == 5].mean()
mean_6 = data.loc[data.index.month == 6].mean()

combined_mean = pd.concat([mean_4, mean_5, mean_6], axis=1)
combined_mean
```

```python
# 以每年每個月為群組的平均
data.groupby(data.index.to_period('M')).mean()
```

```python
# 以每年每個月為群組的詳細統計資料
data.groupby(data.index.to_period('M')).describe()
```

```python
# 以每個月為群組的mean, min, max
ym = data.groupby(data.index.to_period('M')).agg(['mean', 'min', 'max'])
ym
```

```python
# 以每年的同一個月為群組的mean, min, max
data.groupby(data.index.month).agg(['mean', 'min', 'max'])
```

```python
# 以每年每週為群組的詳細統計資料
data.groupby(data.index.to_period('W')).mean()
```

```python
# 以每年的同一週為群組的mean, min, max
data.groupby(data.index.isocalendar().week).agg(['mean', 'min', 'max'])
```
